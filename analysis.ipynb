```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from WRDS / CRSP
df = pd.read_csv("data.csv")
```

```python
# Convert date to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
# Display basic information
print("=== AAPL Stock Analysis (2024) - CRSP/WRDS Data ===")
print(f"Data Period: {df['date'].min().date()} to {df['date'].max().date()}")
print(f"Total Trading Days: {len(df)}")
```

```python
# Calculate daily return and intraday volatility
df['daily_return_pct'] = df['PRC'].pct_change() * 100
df['intraday_volatility'] = (df['ASKHI'] - df['BIDLO']) / df['PRC'] * 100
```

```python
# Display key statistics
print("\n=== Key Statistics ===")
print(f"Average Closing Price: ${df['PRC'].mean():.2f}")
print(f"Highest Price: ${df['PRC'].max():.2f}")
print(f"Lowest Price: ${df['PRC'].min():.2f}")
print(f"Average Daily Volume: {df['VOL'].mean():.0f}")
print(f"Average Daily Return: {df['daily_return_pct'].mean():.2f}%")
```

```python
# Plot 1: Daily Price Trend
plt.figure(figsize=(12, 5))
plt.plot(df['date'], df['PRC'], label='Closing Price')
plt.fill_between(df['date'], df['BIDLO'], df['ASKHI'], alpha=0.2)
plt.title('AAPL Daily Stock Price (2024)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Plot 2: Daily Trading Volume
plt.figure(figsize=(12, 3))
plt.bar(df['date'], df['VOL'], color='green', alpha=0.7)
plt.title('AAPL Daily Trading Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(alpha=0.3)
plt.show()
```

```python
# Plot 3: Daily Return Rate
plt.figure(figsize=(12, 3))
plt.plot(df['date'], df['daily_return_pct'], color='purple')
plt.axhline(0, color='black', linestyle='--')
plt.title('AAPL Daily Return Rate (%)')
plt.xlabel('Date')
plt.ylabel('Return (%)')
plt.grid(alpha=0.3)
plt.show()
```
